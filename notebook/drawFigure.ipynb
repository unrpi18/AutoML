```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4]
y=[1,4,9,16]
plt.plot(x,y)
plt.show()
```

```python
# color:线条颜色，值r表示红色
# marker：点的形状，值o表示点为圆圈标记
#linestyle：线条的形状，值dashed表示用虚线连接各点

plt.plot(x,y,color='r',marker='o',linestyle='dashed')
#axis：坐标轴范围
#语法为axis[xmin,xmax,ymin,ymax]
plt.axis([0,6,0,20])
plt.show()
```

```python
x=[1,2,3,4]
y=[1,4,9,16]

plt.plot(x,y)
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("title")
plt.show()
```
